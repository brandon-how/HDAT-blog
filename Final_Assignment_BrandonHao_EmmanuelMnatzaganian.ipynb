## 1.1. Research Question

Our hospital has been very proactive in analysing data from its Electronic Medical Records (EMR). Through this analysis, we've made some interesting discoveries.

We identified that readmitted patients are not only those who were sicker during their first admission but also those with less support after discharge or those without medical follow-up post-discharge.
Readmitted patients experienced high levels of emotional stress.
Readmitted patients were at a significantly higher risk of acquiring new infections while in the hospital.
Readmissions are highly costly. Patients who were readmitted for the reasons mentioned above tended to be sicker than during their first admission, and their length of stay was significantly longer than during their first admission.

Our hospital has implemented a pilot program named "Redesigning Discharge," which involves a "specialised unit" that coordinates patients' discharges. This pilot program has been running for a year.

This "specialised unit" is composed of a team of doctors, nurses, physiotherapists, and occupational therapists.

The program consists of two main innovations.

1. In the first part of the program, the doctors conduct a new discharge assessment protocol using innovative evidence-based tools to help them decide whether patients are ready to return to their usual place of residence or if it would be more advisable to transfer them to an intermediate or short-term care facility. As such, this program is set to replace the previous protocol concerning where patients should be discharged.

2. The second part of the program includes a group of nurses, physiotherapists, and occupational therapists who visit patients at home or in the intermediate facility after discharge. The frequency of these visits is determined by an assessment conducted prior to discharge, averaging about five visits per patient. Among other things, the team's tasks include ensuring proper wound healing, checking medication compliance, coordinating with a general practitioner or the medical team at the intermediate facility, and assessing the patient's ability to perform basic daily activities such as navigating their environment, using the toilet, and showering. Furthermore, the specialised unit maintains regular telephone contact with patients to check on their well-being. Patients are also encouraged to reach out to the unit if they need any assistance.

This pilot program has been very successful and it has been welcomed by all the stakeholders. Among other things, the program has significantly reduced the number of hospital readmissions. Moreover, the operating costs for this specialised unit have proven to be considerably lower than the costs associated with readmissions

In terms of budget:

1. The average cost of a day in the hospital is 6,000 dollars; readmitted patients tend to stay for an average of 4 days.

2. An intervention by the specialised unit costs on average 2,000 dollars, which includes the cost of the average number of visits per patient, being 5 visits.

3. The cost of an intermediate facility is not considered, as it's covered by Private Health Insurance or other Government Schemes. Additionally, our program is effectively 'donating' a team of medical professionals to these facilities. Therefore, we do not need to account for the cost of these intermediate facilities if a patient is discharged there.

Our hospital is now prepared to implement "Redesigning Discharge" and extend the specialized unit service to all patients at risk of readmission. The challenge we face, however, is that we cannot predict in advance which patients are at risk of readmission. As such, we are currently unable to identify the specific patients we should be targeting

1. Can we develop a machine learning algorithm that can predict the probability of a patient's readmission within 30 days before their discharge?

<b> Rationale: What are you doing to solve this question? - 75 word count limit:</b>

Firstly, we read in the pickle data file as a pandas dataframe.

Secondly, we removed uninformative variables which we will not use in our algorithm. Since we are interested in predicting patients' risk of readmission within 30 days _before_ doctors make their discharge decision, we will remove columns containing discharge information as well as admission and patient IDs.

Lastly, we transformed the response labels so that non-readmissions and readmissions are assigned to 0 and 1 respectively.


```python
# Python code here (15 lines limit, 1 cell limit)
```

```python
# Define file path and read data
data_path = project_path / 'hospital_data_final_assignment.pickle'
hospital = pd.read_pickle(data_path)

# Drop columns not needed in model
hospital_cleaned = hospital.drop(columns = list(hospital.filter(regex='admission_id|patient_id|^discharge')))

# Split the data into predictors and label
X = hospital_cleaned.drop(columns = ['readmission'])
y = [0 if response == 'no' else 1 for response in hospital_cleaned['readmission']]
```

## Question 4: Train and tune the hyperparameters of the Logistic Regression model.

Hyper-parameters:

- `C` values: 10, 100.
- `class_weight`: only two pairs. Choose one of these two combinations and explain why.

    * A weight of (80% for class 1, 20% for class 0) and (70% for class 1, 30% for class 0)
    * A weight of (80% for class 0, 20% for class 1) and (70% for class 0, 30% for class 1).

- `penalty` values: elasticnet, None
- 3-fold cross-validation for the grid search.
- `f1` as the score to choose the best model in the grid search.
- `n_jobs`=-1.
- do not include the heatmaps in the final submission.
- do not change these hyper-parameters.
- keep the remaining set of hyper-parameters in the default state.
---


<b> Rationale: What are you doing to solve this question?-100 word count limit:</b>

Since we are using either `elasticnet` or no penalty, `saga` is the only suitable solver. Since `l1_ratio` is mandatory for `elasticnet`, we set it to 0.5 for equal l1 and l2 penalty. We also unit-variance scale the data as scaling can improve convergence speed and can lead to better model fit. We increased `max_iter` to ensure convergence.

Thereafter, we defined a base logistic regression model with the aforementioned hyperparameters and a hyperparameter grid to search for the best model with 3-fold CV using F1-score. We chose the first `class_weight` combination to assign greater importance to class 1 which is under-represented in our data (17.2%).

Lastly, we fed a pipeline with our scaler and logistic regression to the `GridSearchCV` object and fitted it to the training set.

```python
# Python code here (15 lines limit, and 2 cells limit)
```

```python
# Define data scaler and base logistic regression model
scaler = StandardScaler()
log_mod = LogisticRegression(solver = 'saga', l1_ratio = 0.5, random_state = 30, max_iter = 300)

# Define pipeline for efficient data transformation and model building
pipeline = Pipeline([('Transform', scaler), ('Estimator', log_mod)])

# Define parameter grid for grid searching
log_param_grid = {
    'Estimator__C': [10, 100],
    'Estimator__class_weight': [{0:0.20, 1:0.80}, {0:0.30, 1:0.70}],
    'Estimator__penalty': ['elasticnet', None]
    }
```

```python
%%time
# Define and fit gridsearch object
log_grid_search = GridSearchCV(pipeline, log_param_grid, scoring = 'f1', cv = 3, n_jobs = -1)
log_grid_search.fit(X_train, y_train)
```

If our model performs significantly better in the training set than the test set (low bias, high variance), it suggests potential overfitting. Evaluating the model on the test set tells us how well our model generalises to unseen observations. If test set performance is low (high variance), we may need to re-tune the hyperparameters or try a new algorithm.

To evaluate our model performance, we made class predictions on both training and test sets. We evaluated them against the true labels using confusion matrices and `classification_report` which shows a range of performance metrics.

<b> Rationale: What are you doing to solve this question?-150 word count limit:</b>

Firstly, we defined a random forest classifier model object with random seed 30 for reproducibility. Since the hospital aims to capture as many "readmissions" as possible (i.e., high recall) even if it moderately increases the number of false positives (i.e., reduced precision), we defined a weighted F1-score for choosing the best model in the grid search. The standard F1-score is a harmonic mean of recall and precision with equal weights, but our custom score is a weighted harmonic mean of recall and precision where recall (67%) is weighted twice as heavily as precision (33%).

To assess a broader range of hyperparameters, we conducted a grid search which searched over the sklearn default `max_features` (`2`) and `min_samples_split` (`'sqrt'`) in addition to the `max_features`, `min_samples_split`, `n_estimators` and `class_weight` provided in the question. For the same reasons outlined in Question 4, we chose the first `class_weight` combination.

```python
# Python code here (15 lines limit, and 2 cells limit)
```

```python
# Define random forest object, param grid and scoring strategy for finding best model
rf_mod = RandomForestClassifier(random_state = 30)

rf_param_grid = {
    'n_estimators': [125, 150],
    'class_weight': [{0:0.20, 1:0.80}, {0:0.30, 1:0.70}],
    'max_features': [20, 30, 'sqrt'],
    'min_samples_split': [20, 25, 2]
    }

# Define custom weighted F1-score which gives more weight to recall performance
def weighted_f1(cv_results):
  """
  explanation
  """
  sum_weighted_f1 = np.zeros(cv_results['mean_test_recall'].shape)
  for split in range(0, 3):
    weighted_f1 = 1 / (0.67/cv_results[f'split{split}_test_recall'] + 0.33/cv_results[f'split{split}_test_precision'])
    sum_weighted_f1 += weighted_f1
  mean_weighted_f1 = sum_weighted_f1 / 3
  return np.argmax(mean_weighted_f1)
```

```python
%%time
# Define and fit gridsearch object
rf_grid_search = GridSearchCV(rf_mod, rf_param_grid, cv = 3, n_jobs = -1, scoring = ['recall', 'precision', 'f1'], refit = weighted_f1)
rf_grid_search.fit(X_train, y_train)
```

```python
recall_wt = 0.65
precision_wt = 1 - recall_wt

df = pd.DataFrame(rf_grid_search.cv_results_)
df_perf = df.filter(regex = '^param_|^split|^mean')

df_perf.loc[:, 'weighted_f1_split0'] = 1 / (recall_wt/df_perf.loc[:, 'split0_test_recall'] + precision_wt/df_perf.loc[:, 'split0_test_precision'])
df_perf.loc[:, 'weighted_f1_split1'] = 1 / (recall_wt/df_perf.loc[:, 'split1_test_recall'] + precision_wt/df_perf.loc[:, 'split1_test_precision'])
df_perf.loc[:, 'weighted_f1_split2'] = 1 / (recall_wt/df_perf.loc[:, 'split2_test_recall'] + precision_wt/df_perf.loc[:, 'split2_test_precision'])

df_perf.loc[:, 'weighted_f1_mean'] = (df_perf.loc[:, 'weighted_f1_split0'] + df_perf.loc[:, 'weighted_f1_split1'] + df_perf.loc[:, 'weighted_f1_split2']) / 3

df_perf.loc[df_perf.weighted_f1_mean == max(df_perf.weighted_f1_mean), ['mean_test_recall', 'mean_test_precision']]
```
